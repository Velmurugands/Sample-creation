# Merge DataFrame

```python
temperature_df=pd.DataFrame({
    "city":["mumbai","delhi","bengalore","hyderabad"],
    "temperature":[32,35,30,40]
})
temperature_df
```

```python
humidity_df=pd.DataFrame({
    "city":["delhi","mumbai","bengalore"],
    "humidity":[68,65,75]
})
humidity_df
```

```python
merge_df=pd.merge(temperature_df,humidity_df,on='city')
```

```python
merge_df
```

```python
#outer join
merge_df_dupl=pd.merge(temperature_df,humidity_df,on='city',how='outer')
merge_df_dupl
```

```python
sec1 =pd.DataFrame({"Stuid":[7,8,9,10,11,12],
                    "StuName":['x','y','z','a','b','c'],
                    "Secname":['sec2','sec2','sec2','sec2','sec2','sec2']})
sec1
```

```python
sec2 =pd.DataFrame({"Stuid":[7,8,9],
                    "parentName":['x','y','z'],
                    "mobile":['sec2','sec2','sec2']})
sec2
```

```python
dummy = pd.concat([sec1, sec2],ignore_index=True)
dummy
```

```python
dummy1 = pd.merge(sec1, sec2,on='Stuid')

dummy1
```
